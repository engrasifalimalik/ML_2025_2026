```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(iris.data,iris.target,test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear', C=1.0, gamma='auto')
model.fit(x_train,y_train)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
import numpy as np
avg_score={}
kernel = ['linear', 'poly', 'poly', 'rbf', 'rbf']
C = [1,2,3,4,5]
for kval in kernel:
    for cval in C:
        cv_score = cross_val_score(SVC(kernel=kval, C = cval, gamma='auto'), iris.data,iris.target, cv=5)
        avg_score[kval+' '+str(cval)] = np.average(cv_score)
avg_score
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(SVC(gamma='auto'),{
 'C':[1,2,3,4,5],
'kernel':['linear', 'poly', 'rbf', 'rbf'],
},cv=5,return_train_score=False)
clf.fit(iris.data,iris.target)
# clf.best_params_
# clf.best_score_
clf.cv_results_
```

How to check which model will perform best on iris dataset

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
model_params =({
'logistic_regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C':[1,10,20]
        }
},
'svm':{
        'model':SVC(kernel='linear', C=1.0, gamma='auto'),
        'params':{
            'C':[1,10,20],
        'kernel':['linear', 'poly', 'rbf']
        }
},
'random_forest':{
        'model':RandomForestClassifier(n_estimators=100),
        'params':{
        'n_estimators':[1,10,20]
        }
}
})
```

```python
score = []
for model_name,mp in model_params.items():
    clf = GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(iris.data,iris.target)
    score.append({
        'model_name':model_name,
        'score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
score
```

```python
df = pd.DataFrame(score)
```
