```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('homeprices.csv')
df.head()
```

```python
df.town
```

```python
df_dummies = pd.get_dummies(df.town,dtype = int)
df_dummies.head()
```

```python
df_new = pd.concat([df,df_dummies],axis = 'columns')
```

```python
df_new.head()
```

```python
df_new.drop(['town','west windsor'],axis='columns',inplace = True)
```

```python
df_new.head()
```

```python
x = df_new.drop('price',axis = 1)
y = df_new['price']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict(x)
```

```python
model.predict([[2300,1,0]])
```
