```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris

from utilities.greetings import average

iris = load_iris()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(iris.data, iris.target)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
print("Logistic Regression:",cross_val_score(LogisticRegression(), x_train, y_train, cv=5))
print("Decision Tree:",cross_val_score(RandomForestClassifier(), x_train, y_train, cv=5))
print("Random Forest:",cross_val_score(RandomForestClassifier(), x_train, y_train, cv=5))
print("SVM:",cross_val_score(SVC(), x_train, y_train, cv=5))
```

```python
#Single score of each model using round and mean
print("Logistic Regression:",round(cross_val_score(LogisticRegression(random_state=10), x_train, y_train).mean(),3))
print("Decision Tree:",round(cross_val_score(DecisionTreeClassifier(random_state=10), x_train, y_train).mean(),3))
print("Random Forest:",round(cross_val_score(RandomForestClassifier(n_estimators=5), x_train, y_train).mean(),3))
print("SVM:",round(cross_val_score(SVC(kernel='linear'), x_train, y_train).mean(),3))
```
