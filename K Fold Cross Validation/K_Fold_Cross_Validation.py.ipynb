```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
lr = LogisticRegression(max_iter=200)
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
```

```python
svm = SVC()
svm.fit(x_train,y_train)
svm.score(x_train,y_train)
```

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
```

```python
def performance(model,x_train,x_test,y_train,y_test):
    score = model.score(x_test,y_test)
    return score
```

```python
print('Logistic Regression:',performance(lr,x_train,x_test,y_train,y_test))
print('SVM:',performance(svm,x_train,x_test,y_train,y_test))
print('Random Forest:',performance(rf,x_train,x_test,y_train,y_test))
```

```python
for a in range(1):
    print('Logistic Regression:',performance(lr,x_test,y_test))
    print('SVM:',performance(svm,x_test,y_test))
    print('Random Forest:',performance(rf,x_test,y_test))
```

```python
score_lr = []
score_svm = []
score_rf = []
for train_index, test_index in kf.split(digits.data):
    x_train,x_test,y_train,y_test = train_test_split(digits.data[train_index],digits.target[train_index])
    score_lr.append(performance(lr,x_train,x_test,y_train,y_test))
    score_svm.append(performance(svm,x_train,x_test,y_train,y_test))
    score_rf.append(performance(rf,x_train,x_test,y_train,y_test))
```
