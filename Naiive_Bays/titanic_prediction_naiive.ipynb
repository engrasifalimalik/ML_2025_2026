```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.drop(['Name','Pclass','SibSp','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
x = df.drop('Survived',axis='columns')
y = df['Survived']
```

```python
x.head()
```

```python
dummies = pd.get_dummies(x.Sex,dtype="int")
```

```python
dummies
```

```python
inputs = pd.concat([x,dummies],axis='columns')
```

```python
inputs.head()
```

```python
inputs.drop('Sex',axis='columns',inplace=True)
```

```python
inputs.isnull().any()
```

```python
inputs.Age[2]
```

```python
inputs.Age.fillna(inputs.Age.mean(),inplace=True)
inputs.isnull().any()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(inputs,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB(priors=None,var_smoothing=1e-50)
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
model.predict_proba(x_test)
```
