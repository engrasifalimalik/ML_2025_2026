```python
import pandas as pd
df = pd.read_csv('spam.csv')
df.head()
```

```python
# dummies = pd.get_dummies(df.Category)
# dummies.head()
```

```python
df.groupby('Category').describe()
```

```python
df['spam'] = df['Category'].apply(lambda x:1 if x == "spam" else 0)
df.head()
```

```python
df.drop('Category', axis='columns', inplace=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df['Message'],df['spam'],test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

m = CountVectorizer()
```

```python
x_train_count = m.fit_transform(x_train.values)
x_train_count.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(x_train_count,y_train)
```

```python
email = [
    'Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005'
]
email_test = m.transform(email)
prediction = model.predict(email_test)
print(prediction)
```

```python
x_test_count = m.transform(x_test)
model.score(x_test_count,y_test)
```

Alternate method using Pipeline

```python
from sklearn.pipeline import Pipeline
pipeline = Pipeline([
    ('vector', CountVectorizer()),
    ('nb', MultinomialNB()),
])
```

```python
pipeline.fit(x_train,y_train)
```

```python
pipeline.score(x_test,y_test)
```
